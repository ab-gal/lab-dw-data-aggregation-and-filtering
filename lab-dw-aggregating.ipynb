2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
df2=df2[df2['Response'] == 'Yes']
print(df2['Monthly Premium Auto'].mean())
df_Clv_PolicyType_Gender=df2.pivot_table(index=['Policy Type', 'Gender'], values=['Total Claim Amount', 'Customer Lifetime Value', 'Monthly Premium Auto'], aggfunc='mean')
#calculating
df_Clv_PolicyType_Gender['Cost/Expected Revenue']=((df_Clv_PolicyType_Gender['Total Claim Amount'])/(df_Clv_PolicyType_Gender['Customer Lifetime Value'])*100)
df_Clv_PolicyType_Gender['Breakeven']=((df_Clv_PolicyType_Gender['Total Claim Amount'])/(df_Clv_PolicyType_Gender['Monthly Premium Auto']))
#formating
df_Clv_PolicyType_Gender['Monthly Premium Auto']=df_Clv_PolicyType_Gender['Monthly Premium Auto'].round(2)
df_Clv_PolicyType_Gender['Total Claim Amount']=df_Clv_PolicyType_Gender['Total Claim Amount'].round(2)
df_Clv_PolicyType_Gender['Customer Lifetime Value']=df_Clv_PolicyType_Gender['Customer Lifetime Value'].round(2)
df_Clv_PolicyType_Gender['Cost/Expected Revenue']=df_Clv_PolicyType_Gender['Cost/Expected Revenue'].round(2).apply(lambda x: f'{x}%')
df_Clv_PolicyType_Gender['Breakeven']=df_Clv_PolicyType_Gender['Breakeven'].round(2)
df_Clv_PolicyType_Gender
```

The most profitable and less risk for the company are:
- Corporate policy, female, with an average of 6.33% of costs for the expected revenue; and Personal policy, female, with an av. of 8.52% of costs for the expected revenue.
- Women customers are, in average, less risky.
- Nevertheless, the total claims are equivalent to aproximately 4.8 months of premiums regardless the type and gender.
